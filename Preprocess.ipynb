```python
train = pd.read_csv('Train.csv')
print(train.shape)
train.head()
```

```python
Test = pd.read_csv('Test.csv')
print(Test.shape)
Test.head()
```

```python
game_statistics = pd.read_csv("game_statistics.csv")
```

```python
train["Train"] = 1
Test["Train"]= 0

train_test = pd.concat([train, Test])#for easy manipulation
```

```python
#Merging all data
all_data = game_statistics.merge(train_test, on="Game_ID")
```

```python
home_team_df=all_data[all_data['Team']==all_data['Home Team']].groupby(['Game_ID','Team']).agg("sum")
away_team_df=all_data[all_data['Team']==all_data['Away Team']].groupby(['Game_ID','Team']).agg("sum")
```

```python
home_team_df
```

```python
home_team_df.reset_index(inplace=True)
away_team_df.reset_index(inplace=True)
```

##Feature Engineering 

```python
home_team_df['SoT/shots']=home_team_df['SoT']/home_team_df['Shots']
home_team_df['acc_pass']=home_team_df['Accurate passes']/home_team_df['Passes']
home_team_df['adv SoT/shots']=away_team_df['SoT']/away_team_df['Shots']
home_team_df['adv acc_pass']=away_team_df['Accurate passes']/away_team_df['Passes']
```

```python
home_team_df
```

```python
df=home_team_df.merge(train_test, on="Game_ID")
```

```python
df.drop(['id','X','Y','Start_minutes','End_minutes','next_x','next_y','event_id','next_event_id','xt_value','Season_x','Match_ID_x','Train_x','Team'],axis=1,inplace=True)
```

```python
# time features 
df['created_at'] = pd.to_datetime(df['Date'])
df['year'] = df['created_at'].dt.year
df['year'] = df['year'].astype(float)
df['month'] = df['created_at'].dt.month
df['day'] = df['created_at'].dt.day
df['weekday'] = df['created_at'].dt.weekday
df['weekofyear'] = df['created_at'].dt.weekofyear
```

```python
df.to_csv("Final_df.csv",index=False)
```
